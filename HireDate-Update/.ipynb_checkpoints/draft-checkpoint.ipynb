```python
def rm_space(s: str):
    """
    Remove space at the begin an the end of the given string, if they exist.

    :param s: the string to be modified
    :return: the modified string

    """
    if not isinstance(s, str):
        return s
    return s.strip()
```

```python
# 再从df2员总表中读取信息, 来更新“最早加入公司日期”
# 先对df2所有员工筛选出最早日期和最晚日期
# 首先，找出df3中缺少的df2员工，他们的”当前职位开始日期“为所有日期最晚，“最早加入公司日期”为所有日期最早
# 然后，对df3中原先有的df2员工，他们的”最早加入公司日期“为所有日期最早，“当前职位开始日期”不变
def to_date(datetime):
    """
    Convert datatime type to string.

    :param datetime: datetime object
    :return: string representation of the date
    """
    if datetime == '-':
        return '-'
    else:
        return str(datetime.date())
    
df2["入职日期"] = df2["入职日期"].apply(to_date)
df2["入职日期YG"] = df2["入职日期YG"].apply(to_date)
df2["离职时间"] = df2["离职时间"].apply(to_date)
df2["工资年月"] = df2["工资年月"].apply(to_date)
```
